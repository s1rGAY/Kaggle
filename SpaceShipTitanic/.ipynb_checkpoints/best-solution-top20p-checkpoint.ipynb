# Feature engineering

Ideas:
1. Feature Name is consist of First_Name-Space-Second_Name. Of course, a second name may not show that people with the same second name are from the same family. But the probability that many namesakes was traveled on the same spaceship is too small.
      New feature -> marital status: if two people have the same name = 2, tree = 3, etc.
   
   
   UPD. just use one hot

2. PassengerID is consist of two parts : gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group and (as was explained in data) People in a group are often family members, but not always. So, in my opinion pp is useless and ill make new feature only with gggg.
      New feature -> groupe: string of gggg(with one-hot)

## 1

```python
merged_df['Name'] = merged_df['Name'].str.split(n=1).str[1]
merged_df.head()
```

## 2

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
merged_df_list = merged_df['PassengerId'].str.split('_').str[0].tolist()
merged_df['GroupMemb'] = merged_df['PassengerId']
for i in range(merged_df.shape[0]):
    merged_df['GroupMemb'][i] = merged_df_list.count(merged_df_list[i])
```

```python
merged_df.head()
```

## 3

```python
merged_df['Deck'] = merged_df['Cabin'].str.split('/').str[0]
merged_df['Side'] = merged_df['Cabin'].str.split('/').str[2]
merged_df.drop('Cabin', axis=1, inplace=True)
```

# Data preparation

```python
upd_train = merged_df[merged_df.train_test == 1]
upd_test = merged_df[merged_df.train_test == 0]
```

```python
for feature in upd_train.select_dtypes("number"):
    upd_train[feature] = upd_train[feature].fillna(upd_train[feature].median())
    upd_test[feature] = upd_test[feature].fillna(upd_test[feature].median())
```

```python
upd_merged_df = pd.concat([upd_train,upd_test])
```

```python
upd_merged_df.head()
```

```python
upd_merged_df = pd.get_dummies(upd_merged_df, columns=['HomePlanet','CryoSleep','Destination','VIP', 'Name', 'Deck', 'Side'])
```

```python
upd_merged_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
upd_merged_df_scaled = upd_merged_df.copy()
upd_merged_df_scaled[['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']] = scale.fit_transform(upd_merged_df_scaled[['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']])
```

```python
#back slpit
train = upd_merged_df_scaled[upd_merged_df_scaled.train_test == 1].drop(['train_test'], axis =1)
test = upd_merged_df_scaled[upd_merged_df_scaled.train_test == 0].drop(['train_test'], axis =1)
```

# Models

```python
from sklearn import model_selection
from sklearn.metrics import f1_score
```

```python
train_y = train.Transported.astype(int)
train.drop('Transported', axis=1, inplace=True)
train.drop('PassengerId', axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(train, train_y, test_size=0.33, shuffle=True)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from xgboost import XGBClassifier
from xgboost import XGBRFClassifier
```

```python
#1. GaussianNB
model_gnb = GaussianNB()
model_gnb.fit(X_train, y_train)
print(f1_score(model_gnb.predict(X_test),y_test))
```

```python
#2. LogisticRegression
model_lr = LogisticRegression(penalty='l2' ,max_iter = 1000)
model_lr.fit(X_train, y_train)
print(f1_score(model_lr.predict(X_test),y_test))
```

```python
#3. KNeighborsClassifier
model_knn = KNeighborsClassifier(metric='minkowski')
model_knn.fit(X_train, y_train)
print(f1_score(model_knn.predict(X_test),y_test))
```

```python
#4. SVC
model_svc = SVC(probability=True)
model_svc.fit(X_train, y_train)
print(f1_score(model_svc.predict(X_test),y_test))
```

```python
X_train['GroupMemb'] = X_train['GroupMemb'].astype(float)
X_test['GroupMemb'] = X_test['GroupMemb'].astype(float)
```

```python
#5. XGBClassifier
model_xgbc = XGBClassifier(nthread=-1)
model_xgbc.fit(X_train, y_train)
print(f1_score(model_xgbc.predict(X_test),y_test))
```

```python
#5. XGBRFClassifier
model_xgbrf = XGBRFClassifier(n_estimators = 100)
model_xgbrf.fit(X_train, y_train)
print(f1_score(model_xgbrf.predict(X_test),y_test))
```

```python
# Main model
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import cross_val_score

voting_clf = VotingClassifier(estimators = [('lr',model_lr),
                                            ('knn',model_knn),
                                            ('rf',model_xgbrf),
                                            ('gnb',model_gnb),
                                            ('svc',model_svc),
                                            ('xgbc',model_xgbc)], voting = 'soft')
cv = cross_val_score(voting_clf,X_train,y_train,cv=5)
```

```python
print(cv)
print(cv.mean())
```

```python
voting_clf.fit(X_train,y_train)
```

```python
test['GroupMemb'] = test['GroupMemb'].astype(bool)
```

```python
y_hat_base_vc = voting_clf.predict(test.drop(['Transported','PassengerId'], axis=1)).astype(bool)
basic_submission = {'PassengerId': test.PassengerId, 'Transported': y_hat_base_vc}
base_submission = pd.DataFrame(data=basic_submission)
base_submission.to_csv('/home/siarhei/Programming/sp/base_submission_y_hat_base_vc', index=False)
```

# Improvement


```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
def clf_performance(classifier, model_name):
    print(model_name)
    print('Best Score: ' + str(classifier.best_score_))
    print('Best Parameters: ' + str(classifier.best_params_))
```

```python
#1. GaussianNB
gnb = GaussianNB()
param_grid = {'clf__priors': [None],
              'clf__var_smoothing': [1e-11, 1e-10, 1e-9]}
 
clf_gnb = GridSearchCV(gnb, param_grid = param_grid, cv = 2, verbose = True, n_jobs = -1)
best_clf_gnb = clf_gnb.fit(X_train,y_train)
clf_performance(best_clf_gnb,'GaussianNB')
```

```python
#2. LogisticRegression
lr = LogisticRegression()
param_grid = {'max_iter' : [1000, 1500, 2000],
              'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
              'C' : np.logspace(-4, 4, 30),
              'solver' : ['liblinear', 'newton-cg', 'lbfgs', 'saga']}

clf_lr = GridSearchCV(lr, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_lr = clf_lr.fit(X_train, y_train)
clf_performance(best_clf_lr,'Logistic Regression')
```

```python
#3. KNeighborsClassifier
knn = KNeighborsClassifier()
param_grid = {'n_neighbors' : [3,5,7,9],
              'weights' : ['uniform', 'distance'],
              'algorithm' : ['auto', 'ball_tree','kd_tree'],
              'p' : [1,2]}
clf_knn = GridSearchCV(knn, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_knn = clf_knn.fit(X_train,y_train)
clf_performance(best_clf_knn,'KNN')
```

```python
#4. SVC
svc = SVC(probability = True)
param_grid = tuned_parameters = [{'kernel': ['rbf'], 'gamma': [.1,.5,1,2,5,10],
                                  'C': [.1, 1, 10, 100, 1000]},
                                 {'kernel': ['linear'], 'C': [.1, 1, 10, 100, 1000]},
                                 {'kernel': ['poly'], 'degree' : [2,3,4,5], 'C': [.1, 1, 10, 100, 1000]}]
clf_svc = GridSearchCV(svc, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_svc = clf_svc.fit(X_train,y_train)
clf_performance(best_clf_svc,'SVC')
```

```python
xgbfc = XGBRFClassifier()

param_grid={'min_child_weight': [1, 5, 10],
            'gamma': [0.5, 1, 1.5, 2, 5],
            'subsample': [0.6, 0.8, 1.0],
            'colsample_bytree': [0.6, 0.8, 1.0],
            'max_depth': [3, 4, 5, 10]}

clf_xgbfc = GridSearchCV(xgbfc, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_xgbfc = clf_xgbfc.fit(X_train,y_train)
clf_performance(best_clf_xgbfc,'XGBFC')
```

```python
best_lr = best_clf_lr.best_estimator_
best_knn = best_clf_knn.best_estimator_
best_svc = best_clf_svc.best_estimator_
best_xgbfc = best_clf_xgbfc.best_estimator_
```

```python
best_lr = best_clf_lr.best_estimator_
best_knn = best_clf_knn.best_estimator_
best_svc = best_clf_svc.best_estimator_
best_xgbfc = best_clf_xgbfc.best_estimator_

voting_clf_hard = VotingClassifier(estimators = [('knn',best_knn),('rf',best_xgbfc),('svc',best_svc)], voting = 'hard') 
voting_clf_soft = VotingClassifier(estimators = [('knn',best_knn),('rf',best_xgbfc),('svc',best_svc)], voting = 'soft') 
voting_clf_all = VotingClassifier(estimators = [('knn',best_knn),('rf',best_xgbfc),('svc',best_svc), ('lr', best_lr)], voting = 'soft') 
voting_clf_xgbfc = VotingClassifier(estimators = [('knn',best_knn),('rf',best_xgbfc),('svc',best_svc),('lr', best_lr)], voting = 'soft')

print('voting_clf_hard :',cross_val_score(voting_clf_hard,X_train,y_train,cv=5))
print('voting_clf_hard mean :',cross_val_score(voting_clf_hard,X_train,y_train,cv=5).mean())

print('voting_clf_soft :',cross_val_score(voting_clf_soft,X_train,y_train,cv=5))
print('voting_clf_soft mean :',cross_val_score(voting_clf_soft,X_train,y_train,cv=5).mean())

print('voting_clf_all :',cross_val_score(voting_clf_all,X_train,y_train,cv=5))
print('voting_clf_all mean :',cross_val_score(voting_clf_all,X_train,y_train,cv=5).mean())

print('voting_clf_xgb :',cross_val_score(voting_clf_xgbfc,X_train,y_train,cv=5))
print('voting_clf_xgb mean :',cross_val_score(voting_clf_xgbfc,X_train,y_train,cv=5).mean())
```
