## **<span style="color:#F26300;">4.1 Outliers</span>**

Using below function, I'll detect rows with n number of outliers, I'll drop rows with more than 5 outliers

```python
from collections import Counter
def outlier_detect(df,n,cols):
    rows,to_drop=[],[]
    for col in cols:
        Q1=np.nanpercentile(df[col],25)
        Q3=np.nanpercentile(df[col],75)
        IQR=Q3-Q1
        outlier_point=1.5*IQR
        rows.extend(df[(df[col]<Q1-outlier_point)|(df[col]>Q3+outlier_point)].index)
    for r,c in Counter(rows).items():
        if c>=n: to_drop.append(r)
    return to_drop
```

```python
to_drop=outlier_detect(train_set,5,train_set.select_dtypes('float').columns)
```

```python
train_set.drop(to_drop,inplace=True,axis=0)
```

## **<span style="color:#F26300;">5.7 Chi2 test</span>**

**Let's do a chi2 test to see if there are any features with a pvalue higher than 0.05**

```python
from scipy.stats import chi2_contingency
def chi2_calc(df,target):
    scores=[]
    for col in df.columns:
        ct=pd.crosstab(df[col],target)
        stat,p,dof,expected=chi2_contingency(ct)
        scores.append(p)
    return pd.DataFrame(scores, index=df.columns, columns=['P value']).sort_values(by='P value')
```

```python
chi2_calc(train_set.select_dtypes(['object','bool']),train_set.Transported)
```

**So all the categorical features have significant relevance to the target (pvalue <0.05)**

**Note: OrdinalEncoder (without defining an order) is like LabelEncoder, but we can apply it to multiple columns at once
whereas LabelEncoder, is for one column transformation. The order in which they encode, is ascending, meaning A will be 1 and B will be 2**

```python
#Importing Models
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,GradientBoostingClassifier,AdaBoostClassifier,ExtraTreesRegressor
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

########################################################################
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer,KNNImputer,SimpleImputer
from category_encoders import MEstimateEncoder,PolynomialEncoder,BackwardDifferenceEncoder,LeaveOneOutEncoder,QuantileEncoder,WOEEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import PowerTransformer,QuantileTransformer
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
random_state=0

#setting up the imputer for numerical features
NImputer=IterativeImputer(estimator=ExtraTreesRegressor(max_depth=7,min_samples_leaf=200,random_state=random_state),random_state=random_state,tol=1e-5,max_iter=20)

#setting up the imputer for categorical features (We can explore with this imputer too, I'll let it be here just in case)
CImputer=IterativeImputer(estimator=ExtraTreesClassifier(max_depth=10,min_samples_leaf=100,random_state=random_state),random_state=random_state,tol=1e-5,max_iter=20)

#target encoder: it's usually used for high cardinality features. 
# t_encoder = MEstimateEncoder(m=10, random_state=random_state,handle_missing='return_nan')

#PolynomialEncoder
# p_ecnoder=PolynomialEncoder()

#BackwardDifferenceCoding
# b_encoder=BackwardDifferenceEncoder()

#QuantileEncoder
# q_encoder=QuantileEncoder(m=12,handle_missing='return_nan')

#Weight of Evidence Encoder
we=WOEEncoder(drop_invariant=True,handle_missing='return_nan')


#creating log function that has fit and fit_transform methods because numerical columns are mostly skewed
# def log_transform(x):
#     return np.log(x + 1)
# log_pip=FunctionTransformer(log_transform)

#Using power transformer class for standardizing data distirbution in numerical columns
# pt=PowerTransformer()

#Using quantile transformer
# qt=QuantileTransformer(n_quantiles=500, random_state=0,output_distribution='uniform')

#Using kbinsdiscretizer for skewed numerical features
# kbins = KBinsDiscretizer(n_bins=2,encode='ordinal', strategy='uniform')

#creating two preprocessing pipelines for categorical and numerical data types
numeric_pip=Pipeline(steps=[('NImputer',SimpleImputer(strategy='median')),('Scale',RobustScaler(unit_variance=True))])

nomcat_pip=Pipeline(steps=[('WeightOfEvidence',we),('CImputer',SimpleImputer(strategy='mean'))])

ordcat_pip=Pipeline(steps=[('CImputer',SimpleImputer(strategy='mean'))])

#creating a column transformer to implement the transformation
ct=ColumnTransformer(transformers=[('num',numeric_pip,num),('nomcat',nomcat_pip,nominal_cat),('ordcat',ordcat_pip,ordinal_cat)])
```

```python
#This is for testing to see if the columntransformer works properly
tst = Pipeline([
('coltrns',ct), #COLUMN TRANSFORMER
])
xx=tst.fit_transform(train_set,y)
```

## **<span style="color:#F26300;">6.1 Models and Scoring</span>**

```python
classifiers = []
classifiers.append(RandomForestClassifier(random_state=random_state))
classifiers.append(GradientBoostingClassifier(random_state=random_state)) 
classifiers.append(XGBClassifier(random_state=random_state))
classifiers.append(LGBMClassifier(random_state=random_state))
classifiers.append(CatBoostClassifier(random_state=random_state,verbose=0))
# CREATING A FOR LOOP FOR SCORING EACH MODEL
cv_results = []
cv = KFold(n_splits=5,shuffle=True, random_state=random_state)
for classifier in classifiers :
    classif = Pipeline([
('coltrns',ct), #COLUMN TRANSFORMER
('classifier', classifier)])
    cvs=cross_val_score(classif, train_set, y, scoring = "accuracy", cv = cv, n_jobs=-1)
    cv_results.append(cvs)

cv_means = []
cv_std = []
for cv_result in cv_results:
    cv_means.append(cv_result.mean())
    cv_std.append(cv_result.std())

#CREATING A DATAFRAME OF MODEL SCORES
cv_res = pd.DataFrame({"CrossValMeans":cv_means,"CrossValSDs": cv_std,"Algorithm":["RandomForestClassifier",
                                                                                 "GradientBoostingClassifier",
                                                                                  "XGBClassifier",
                                                                                  "LGBMClassifier",
                                                                                  "CatBoostClassifier",
                                                                                  ]})
#PLOTTING
g = sns.barplot(x="CrossValMeans",y="Algorithm",data = cv_res.sort_values(by='CrossValMeans'), palette="twilight_shifted_r",**{'xerr':cv_std})
g.set_xlabel("Mean Accuracy")
g = g.set_title("Cross validation scores")
g=sns.set(rc={'figure.figsize':(5,5)})

cv_res.sort_values(by='CrossValMeans')
```

## **<span style="color:#F26300;">7. Learning Curve</span>**

Here we're going to take a look at the learning curve, using below function, before tuning the models.

```python
from sklearn.model_selection import learning_curve
def plot_learning_curve(my_model, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=[np.linspace(.1, 1.0, 5)]):
    """Generate a simple plot of the test and training learning curve"""
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        my_model, X, y, cv=cv, n_jobs=-1, train_sizes=train_sizes,scoring="accuracy")
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)


    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    plt.grid()
    return plt
```

```python
for i in [1,2,3,4]: #Index of the last 4 highest scores
    model = Pipeline([
        ('coltrns',ct), #COLUMN TRANSFORMER
        ('classifier', classifiers[i])])
    plot_learning_curve(model,model[1],train_set,y)
```

The lesser the training curve changes, the more the model is overfit. because it's working well on the training data and learning it amazingly, but when it comes to the validation set, it can not generalize and performs poorly.

## **<span style="color:#F26300;">9. Transforming and Predicting Test Set</span>**


```python
#Creating Rel Feature For Test Set
test_set[['Name','Last']] = test_set.Name.str.split(" ", expand=True)
test_set.drop('Name',inplace=True,axis=1)
test_set['Last'].ffill(axis=0,inplace=True) # I'll drop 'Last', becuase I still don't know how to use names to improve the score

#Creating cabin capacity using the number of people in it
tcabin_cap= test_set.Cabin.value_counts()
test_set['CabinCap']=test_set['Cabin'].map(tcabin_cap)
test_set['CabinCap']=test_set['CabinCap'].astype('object')

#Creating cabin deck and cabin side features
test_set[['CabinDeck','CabinSide']]=test_set.Cabin.str.split('/',expand=True)[[0,2]]
test_set['GpId']=test_set.PassengerId.str.split('_',expand=True)[0]
test_set['GpSize']=test_set['GpId'].map(test_set['GpId'].value_counts())
test_set['GpSize']=test_set['GpSize'].astype('object')

#Total expense
test_set[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']]=test_set[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].fillna(0,axis=0)
test_set["TotalExp"] = test_set[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']].sum(axis=1)

for col in ['FoodCourt','Spa','VRDeck','ShoppingMall']:
    colname='%'+col
    test_set[colname]=[0 if test_set['TotalExp'][i]==0 else
                        (test_set[col][i]/test_set['TotalExp'][i])*100 for i in test_set.index]
#Expense ratio based on pairplot
test_set['SPFC']=[0 if test_set['FoodCourt'][i]==0 else
                    (test_set['Spa'][i]/test_set['FoodCourt'][i])*100 for i in test_set.index]
test_set['VRFC']=[0 if test_set['FoodCourt'][i]==0 else
                    (test_set['VRDeck'][i]/test_set['FoodCourt'][i])*100 for i in test_set.index]
#Age
bins= [0,18,40,80]
labels = [1,2,3]
test_set['AgeGp']=pd.cut(test_set['Age'], bins=bins, labels=labels,include_lowest=True)

#H-D
test_set['H-D']=test_set['HomePlanet']+"-"+test_set["Destination"]


#dropping PassengerId

test_set.drop(['ShoppingMall','FoodCourt'],inplace=True,axis=1)

test_set.drop(['Age','Cabin','Last','GpId','PassengerId'],inplace=True,axis=1)

test_set.drop(['Destination','HomePlanet'],axis=1,inplace=True)
```
